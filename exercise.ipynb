# 贝叶斯决策

```python
import numpy as np
'''
P:先验概率
p:x的条件概率
loss:风险矩阵
'''
def Bayes(P,p,loss=[]):
    P=np.array(P)
    p=np.array(p)
    postp=P*p/(sum(P*p))
    if loss==[]:
        '''
        最小错误率决策
        '''
        return postp.argmax(),postp
    else:
        '''
        最小风险决策
        '''
        ans=[]
        for i in range(len(postp)):
            s=0
            for j in range(len(postp)):
                s+=loss[i][j]*postp[j]
            ans.append(s)
        ans=np.array(ans)
        return ans.argmin(),ans
```

## 例子 判断细胞正常与否

正常细胞P($\omega_1$)=0.9 异常细胞P($\omega_2$)=0.1

p(x|$\omega_1$)=0.2,p(x|$\omega_2$)=0.4

```python
labels=["正常细胞","异常细胞"]
loss=[[0,6],[1,0]]
ans,ls=Bayes([0.9,0.1],[0.2,0.4])
print(labels[ans])
ans,ls=Bayes([0.9,0.1],[0.2,0.4],loss)
print(labels[ans])
```

# 人工神经网络

```python
import random
from dataset import *
def sigmoid(x):
    return 1/(1+np.exp(-x))
def sigmoidg(x):
    return np.exp(-x)/(pow(1 + np.exp(-x),2))
def mse(Y,y):
    return sum((Y-y)*(Y-y))/2
class Net:
    def __init__(self,n,m,k):
        self.n=n
        self.m=m
        self.k=k
        self.w1=np.random.normal(0,1,(n,m))
        self.X0=np.zeros(m)
        self.X1=np.zeros(m)
        self.X2=np.zeros(k)
        self.b1=0.1
        self.w2=np.random.normal(0,1,(m,k))
        self.b2=0.1
        self.Y=np.zeros(k)
    def forward(self):
        #输入层-隐藏层
        for i in range(self.m):
            s=self.b1
            for j in range(self.n):
                s+=self.w1[j][i]*self.X[j]
            self.X0[i]=s
        self.X1=sigmoid(self.X0)
        #隐藏层-输出层
        for i in range(self.k):
            s=self.b2
            for j in range(self.m):
                s+=self.w2[j][i]*self.X1[j]
            self.X2[i]=s
        self.Y=sigmoid(self.X2)
    def backward(self):
        loss1=self.Y-self.y
        hide_g=self.Y-self.y
        for k in range(self.k):
            loss1[k]*=sigmoidg(self.X2[k])
            hide_g[k]=loss1[k]*self.X1[k]
        input_g=[0]*self.m
        for m in range(self.m):
            for k in range(self.k):
                input_g[m]+=loss1[k]*self.w2[m][k]
            input_g[m]*=sigmoidg(self.X0[m])
        for k in range(self.k):
            for m in range(self.m):
                self.w2[m][k]-=self.lr*hide_g[k]*self.X1[m]
        for m in range(self.m):
            for n in range(self.n):
                self.w1[n][m]-=self.lr*input_g[m]*self.X[n]
        self.loss+=mse(self.Y,self.y)
    def step(self,X,y,lr):
        self.loss=0
        for i in range(len(X)):
            self.X=X[i]
            self.y=y[i]
            self.lr=lr
            self.forward()
            self.backward()
    def pred(self,X):
        ans=[]
        for i in range(len(X)):
            self.X=X[i]
            self.forward()
            ans.append(self.Y)
        return ans
X,y,X_test,y_test=getData1(0.8)
net=Net(6,4,1)
epoch=10000
for i in range(epoch):
    net.step(X,y,0.1)
    loss=net.loss
    if loss<0.01:
        break
    print("epoch %d"%i,"loss:",loss)
net.pred(X),y
```

# 聚类

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
from sklearn import svm
from dataset import getData1
import random
n=6 #输入的特征数量
m=6 #实际使用的特征数量
X,y,X_test,y_test=getData1(1)
def kmeans(X,cluster_num,max_epoch=100):
    X=np.array(X)
    def calcu_dist(x1,x2):
        return np.sqrt(sum((x1-x2)*(x1-x2)))
    def choose_random_ls(num):
        ans=[]
        for i in range(cluster_num):
            r=random.randint(0,num)
            while(r in ans):
                r=random.randint(0,num)
            ans.append(r)
        return ans
    def judge(x1,x2):
        for i in range(len(x1)):
            if list(x1[i])!=list(x2[i]):
                return False
        return True
    num=len(X)
    center_ls=choose_random_ls(num-1)
    center_ls=np.array([X[i] for i in center_ls])
    epoch=0
    while True:
        clusters=[[] for i in range(cluster_num)]
        for i in range(len(X)):
            mindist=100000000
            mini=-1
            for j in range(len(center_ls)):
                d=calcu_dist(center_ls[j],X[i])
                if d < mindist:
                    mini=j
                    mindist=d
            clusters[mini].append(i)
        new_center_ls=[]
        for i in range(len(center_ls)):
            idls=clusters[i]
            x=X[idls].mean(axis=0)
            new_center_ls.append(x)
        new_center_ls=np.array(new_center_ls)
        if judge(list(new_center_ls),list(center_ls)) or epoch>=max_epoch:
            break
        center_ls=new_center_ls
        if epoch%10==0:
            print("epoch%d done"%epoch)
        epoch+=1
    return center_ls
center_ls=kmeans(X,3)
center_ls
```
